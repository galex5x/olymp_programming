```python
# 1 2 3 4 5 ->
# 2

x = input()
seq = list(map(int, x.split(' ')))
k = int(input())

best = 0
for i in range(0, len(seq) - k + 1): 
    cur_sum = 0
    for j in range(k):
        cur_sum += seq[j + i]
        
    # update max 
    if cur_sum > best:
        best = cur_sum
        
print(best)
```

```python
# 1 2 3 4 5 ->
# 2

x = input()
seq = list(map(int, x.split(' ')))
k = int(input())

prev_sum = 0
best = 0

for i in range(0, len(seq) - k + 1): 
    cur_sum = 0
    
    if i == 0 or i >= len(seq) - k -1:
        for j in range(k):
            cur_sum += seq[j + i]
    else:
        prev_sum -= seq[i - 1]
        prev_sum += seq[i + k - 1]
        cur_sum = prev_sum
    
    # update max 
    if cur_sum > best:
        best = cur_sum
        
    prev_sum = cur_sum
    
print(best)
```

```python
x = input()
seq = list(map(int, x.split(' ')))
flag = 'false'

def check_order(s):
    """
    Check for strong order
    """
    for i in range(0,len(s)-1):
        if s[i] >= s[i+1]:
            return False
        
    return True
    
for i in range(len(seq)):
    if check_order(seq[0:i] + seq[i + 1 : len(seq)]) == True:
        flag = 'true'
        break
        
print(flag)
```

```python
up_speed = int(input())
down_speed = int(input())
desiredHeight = int(input())

size = 0
days_count = 0

while size != desiredHeight:
    size += up_speed
    days_count += 1
    if size >= desiredHeight:
        break
    size -= down_speed
    
print(days_count)
```

```python
s1 = input()
s2 = input()

ans = 0

def freq(s):
    d = {}
    for a in s:
        if a in d:
            d[a] += 1
        else:
            d[a] = 1
            
    return d

d1 = freq(s1)
d2 = freq(s2)

for (key, value) in d1.items():
    if key in d2:
        ans += min(d2[key], value)

print(ans)
```

```python
s = input()

max_len = 0
str_ans = ''
max_str_ans = ''
cur_len = 0

for i in range(len(s)):
    
    if ord(s[i]) - ord('0') >= 0 and ord(s[i]) - ord('0') <= 9:
        cur_len += 1
        str_ans += s[i]
    else:
        if cur_len > max_len:
            max_len = cur_len
            max_str_ans = str_ans
        cur_len = 0
        str_ans = ''

if cur_len > max_len:
    max_len = cur_len
    max_str_ans = str_ans  
    
print(max_str_ans)
```

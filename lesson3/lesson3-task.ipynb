```python
n = int(input())
a = int(input())
b = int(input())

# A * [1] + B * [3] total 
total_count = 3 * b + a

number_of_complects = 0

while(True):  
    if a < 0:
        break 
        
    part_3 = n // 3  # desired count of 3-complects
    
    put_3 = min(part_3, b) # may push 3-complects
    
    put_1 = n - put_3 * 3   # may push 1-complects
    
    if put_1 <= a and put_3 <= b:
        number_of_complects += 1
        a -= put_1
        b -= put_3
    else:
        break
    

print(number_of_complects)
```

```python
a = int(input()) 
b = int(input())
c = int(input()) 

# 5, 7, 11 -> 9,3 

# 4,5,6 -> 1) (7,4)
#          2) (3,1)


# 5, 5 + 2, 5 + 4, 5 + 6.

delta_ba = b - a 
delta_cb = c - b

if delta_cb == delta_ba:
    print(a - delta_ba, 1)
    
elif delta_cb == 2 * delta_ba:
    print((c + b) // 2, 3)
else:
    print((a + b) // 2, 2)
```

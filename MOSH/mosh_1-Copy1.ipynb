```python
input_str = list(input().split(' '))
n = int(input_str[0])
m = int(input_str[1])


def get_all_factors(x):
    all_factors = []
    for i in range(1, x + 1):
        if x % i == 0:
            all_factors.append(i)
    return all_factors


def factorization(list_all_factors, n):
    
    answer = 0
    for j in range(len(list_all_factors) - 1):
        result = list_all_factors[j]
        for i in range(j+1, len(list_all_factors)):
            result *= list_all_factors[i]
            if result > n:
                break
            elif result == n:
                answer += 1
            
    return answer 
                           
print(factorization(get_all_factors(n), n))
```

### deque

```python
from collections import deque 
      
# Declaring deque 
queue = deque(['Anton','Timur','Sasha'])  
      
print(queue)
```

- append():- Эта функция используется для вставки значения в аргументе в правый конец deque.
- appendleft():- Эта функция используется для вставки значения в аргументе в левый конец deque.

```python
import collections

de = collections.deque([1, 2, 3])
print("deque: ", de)
  
de.append(4)
  
print("\nДек после вставки справа : ")
print(de)
  
de.appendleft(6)
  
print("\nДек после вставки слева : ")
print(de)
```

```python
import collections
  
de = collections.deque([6, 1, 2, 3, 4])
print("deque: ", de)

de.pop()
  
print("\nПосле удаления справа : ")
print(de)
  
de.popleft()
  
print("\nПосле удаления слева : ")
print(de)
```

```python
import collections
  
de = collections.deque([1, 2, 3, 3, 4, 2, 4])

print ("Число 4 стоит на позиции : ")
print (de.index(4,2,5)) # ищем в диапазоне от 2-ого до 5-ого
  
de.insert(4,3)
  
print ("После вставки 3 на 5-ю позицию : ")
print (de)
  
print ("Количество троек в деке : ")
print (de.count(3))
  
de.remove(3)
  
print ("После удаления первой встретившейся тройки : ")
print (de)
```

```python
import collections
  
de = collections.deque([1, 2, 3,])
  
de.extend([4,5,6])
  
print ("После расширения на конце : ")
print (de)
  
de.extendleft([7,8,9])

print ("После расширения в начале : ")
print (de)
  
de.rotate(-3)
  
print ("После сдвига на 3 позиции влево : ")
print (de)
  
de.reverse()
  
print ("После зеркального отображения : ")
print (de)
```
